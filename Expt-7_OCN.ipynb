```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
f1=0.025
M1=(f1/(3*10**5)*850)
f2=0.004
M2=(f2/(3*10**5)*1300)
```

```python
sig_lam=np.arange(10,100,10)
sig_m1=sig_lam*M1
sig_lam1=np.arange(10,100,10)
sig_m2=sig_lam*M2
```

```python

plt.title("RMS Pulse Brodening")
plt.xlabel("Sigma Lambda")
plt.ylabel("RMS")
plt.plot(sig_lam,sig_m1,color='r',label='850nm')
plt.plot(sig_lam,sig_m2,color='g',label='12000nm')
plt.legend()
plt.show
print("End Code")
```
